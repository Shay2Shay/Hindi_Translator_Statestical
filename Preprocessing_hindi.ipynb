# Data Loading Hindi

```python
import pandas as pd
```

```python
data = pd.read_csv('./hindi_english_parallel.csv')
data.tail()
```

```python
hindi = data['hindi']
print(hindi)
```

### Filter Hindi words lenght > 100

```python
SENETNCE_CHAR_LEN = 100
a = []
for i in hindi:
    if type(i) == float or len(i) <= SENETNCE_CHAR_LEN:
        a.append(False)
    else:
        a.append(True)
hindi = hindi[a]
```

```python
hindi
```

### Filter only hindi words

```python
english_vocab = 'abcdefghijklmnopqrstuvwxyz'
english_vocab += english_vocab.upper()
english_vocab = list(set(english_vocab) - set('xviXVI'))
```

```python
def onlyHindi(sentence : str, english_vocab = english_vocab):
    for i in english_vocab:
        if i in sentence:
            return False
    return True

filteredEnglishData = [onlyHindi(str(sen)) for sen in hindi]
print("============================  SELECTED  ========================")
print(hindi[filteredEnglishData][-10:])
print("============================  REJECTED  ========================")
print(hindi[map(lambda x : not x, filteredEnglishData)][-19:-9])

hindi = hindi[filteredEnglishData]
```

```python
HindiTokns = []
for index, i in enumerate(hindi):
    HindiTokns.append(i.split(" "))
```

# Stemming

```python
suffixes = {
    1: [u"ो",u"े",u"ू",u"ु",u"ी",u"ि",u"ा"],
    2: [u"कर",u"ाओ",u"िए",u"ाई",u"ाए",u"ने",u"नी",u"ना",u"ते",u"ीं",u"ती",u"ता",u"ाँ",u"ां",u"ों",u"ें"],
    3: [u"ाकर",u"ाइए",u"ाईं",u"ाया",u"ेगी",u"ेगा",u"ोगी",u"ोगे",u"ाने",u"ाना",u"ाते",u"ाती",u"ाता",u"तीं",u"ाओं",u"ाएं",u"ुओं",u"ुएं",u"ुआं"],
    4: [u"ाएगी",u"ाएगा",u"ाओगी",u"ाओगे",u"एंगी",u"ेंगी",u"एंगे",u"ेंगे",u"ूंगी",u"ूंगा",u"ातीं",u"नाओं",u"नाएं",u"ताओं",u"ताएं",u"ियाँ",u"ियों",u"ियां"],
    5: [u"ाएंगी",u"ाएंगे",u"ाऊंगी",u"ाऊंगा",u"ाइयाँ",u"ाइयों",u"ाइयां"],
} #suffixes tin hindi language that change tense, gender, number but not the meaning
```

```python
! pip install tqdm
from tqdm import tqdm

stems=list()
for line in tqdm(HindiTokns):
    for word in line:
        for L in range(1,5): 
            if len(word) >= L + 1:
                for suffix in suffixes[L]:
                    if word.endswith(suffix):
                        word=word[:-L] #stripping the suffix from the word
        if word:
            stems.append(word)
```

```python
all_stems = set()
for i in tqdm(stems):
    all_stems.add(i)
```
